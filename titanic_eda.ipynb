## 0402 Explorar Nulos
Cuando sea posible es recomendable entender la causa de los nulos. Hay algún patrón que se repite entre ellos?

Veámoslo para cada una de las variables con nulos:

```
# Exploring 'Age' nulls
print("\n--- Exploring 'Age' Nulls ---")
print("Number of nulls in Age:", df['Age'].isnull().sum())
print("Median Age:", df['Age'].median())
print("Median Age by Sex:", df.groupby('Sex')['Age'].median())
print("Median Age by Pclass:", df.groupby('Pclass')['Age'].median())
print("Median Age by Pclass and Sex:\n", df.groupby(['Pclass', 'Sex'])['Age'].median())

# Exploring 'Cabin' nulls
print("\n--- Exploring 'Cabin' Nulls ---")
print("Number of nulls in Cabin:", df['Cabin'].isnull().sum())
# Create a new column to indicate if a Cabin value is present
df['Has_Cabin'] = df['Cabin'].apply(lambda x: 0 if pd.isnull(x) else 1)
print("Survival rate for passengers with and without Cabin information:\n", df.groupby('Has_Cabin')['Survived'].mean())

# Exploring 'Embarked' nulls
print("\n--- Exploring 'Embarked' Nulls ---")
print("Number of nulls in Embarked:", df['Embarked'].isnull().sum())
print("Rows with null 'Embarked' values:\n")
display(df[df['Embarked'].isnull()])
print("Most frequent 'Embarked' port:", df['Embarked'].mode()[0])

# Drop the temporary 'Has_Cabin' column
df.drop(columns=['Has_Cabin'], inplace=True, errors='ignore')
```

## 0403 Tratamiento de Nulos
Una vez hemos detectado los nulos presentes en la tabla podemos corregirlos de varias formas.

### 040301 Corrección de nulos 'Age'
No parece haber una causa evidente para que algunas edades no estén imputadas. Es una variable numérica que debemos rellenar también con valores numéricos.
De más sencillo a más complejo, podríamos:


1.   Rellenar los nulos con la mediana de los valores de edad no nulos
2.   Rellenar los nulos de las mujeres con la mediana de edad de éstas, y los hombres con la suya
3.   Rellenar los nulos sirviéndonos del título en el nombre, utilizando la mediana del resto de pasajeros con el mismo título.



```
# 1. Rellenar los nulos con la mediana de los valores de edad no nulos

df_copy_median = df.copy()
median_age_overall = df_copy_median['Age'].median()
df_copy_median['Age_filled_overall_median'] = df_copy_median['Age'].fillna(median_age_overall)

print(f"Mediana global de la edad: {median_age_overall}")
print("Valores nulos en 'Age' después de la imputación (mediana global):", df_copy_median['Age_filled_overall_median'].isnull().sum())
```

```
# 2. Rellenar los nulos de las mujeres con su mediana y los hombres con la suya

df_copy_median_sex = df.copy()
df_copy_median_sex['Age_filled_by_sex_median'] = df_copy_median_sex.groupby('Sex')['Age'].transform(lambda x: x.fillna(x.median()))

print("Mediana de edad por sexo:")
display(df_copy_median_sex.groupby('Sex')['Age'].median())
print("Valores nulos en 'Age' después de la imputación (mediana por sexo):", df_copy_median_sex['Age_filled_by_sex_median'].isnull().sum())

# 3. Rellenar los nulos sirviéndonos del título en el nombre

df_copy_median_title = df.copy()
# Extraer el título del nombre
df_copy_median_title['Title'] = df_copy_median_title['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
# Mapear títulos menos comunes a 'Rare' y estandarizar otros
df_copy_median_title['Title'] = df_copy_median_title['Title'].replace(['Lady', 'Countess','Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona', 'Mlle', 'Ms', 'Mme'], ['Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Miss', 'Miss', 'Mrs'])

# Rellenar los nulos de 'Age' con la mediana según el 'Title'
df_copy_median_title['Age_filled_by_title_median'] = df_copy_median_title.groupby('Title')['Age'].transform(lambda x: x.fillna(x.median()))

print("\nMediana de edad por título:")
display(df_copy_median_title.groupby('Title')['Age'].median())
print("Valores nulos en 'Age' después de la imputación (mediana por título):", df_copy_median_title['Age_filled_by_title_median'].isnull().sum())

# Puedes elegir cuál de estas columnas usarás o si quieres mantener la original por ahora.
# Por simplicidad, aquí solo mostramos el resultado de cada método.
```

### 040302 Corrección de nulos 'Embarked'
Solo hay dos nulos aquí, así que vamos a eliminar estas dos filas.

```
# Eliminar las filas donde 'Embarked' es nulo
df.dropna(subset=['Embarked'], inplace=True)
print("Valores nulos en 'Embarked' después de la corrección:")
display(df['Embarked'].isnull().sum())
```

```
# En cuanto a la columna 'Cabin', hay muchos nulos. Es además una
# columna no numérica, así que vamos a dejarlos sin tocar al menos de momento.
# Lo rellenamos con 'indeterminado', respetando el tipo de dato
df['Cabin'] = df['Cabin'].fillna('indeterminado')
print("Valores nulos en 'Cabin' después de la corrección:")
display(df['Cabin'].isnull().sum())
```

### Imputación de 'Age' en el DataFrame principal (`df`)

Para los siguientes análisis, imputaremos los valores nulos de 'Age' en nuestro `df` principal utilizando la mediana según el título de cada pasajero, como exploramos anteriormente.

```
# Extraer el título del nombre y mapearlos
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
df['Title'] = df['Title'].replace(['Lady', 'Countess','Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona', 'Mlle', 'Ms', 'Mme'], ['Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Miss', 'Miss', 'Mrs'])

# Rellenar los nulos de 'Age' con la mediana según el 'Title'
df['Age'] = df.groupby('Title')['Age'].transform(lambda x: x.fillna(x.median()))

# Verificar que no queden nulos en 'Age' y que la columna 'Title' se haya añadido
print("Valores nulos en 'Age' después de la imputación:", df['Age'].isnull().sum())
display(df[['Name', 'Title', 'Age']].head())
```

### Qué % fallece entre los 50 que pagaron más por su billete y los 50 que menos?

```
# 50 pasajeros que pagaron más por su billete
top_50_fare = df.nlargest(50, 'Fare')
percentage_died_top_50 = (top_50_fare['Survived'] == 0).mean() * 100
print(f"Porcentaje de fallecidos entre los 50 que pagaron más: {percentage_died_top_50:.2f}%")
```

```
#pasajeros que pagaron menos por su billete
bottom_50_fare = df.nsmallest(50, 'Fare')
percentage_died_bottom_50 = (bottom_50_fare['Survived'] == 0).mean() * 100
print(f"Porcentaje de fallecidos entre los 50 que pagaron menos: {percentage_died_bottom_50:.2f}%")
```
